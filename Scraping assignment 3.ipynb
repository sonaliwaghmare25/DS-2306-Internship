5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import regex as re
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.get('https://www.google.co.in/maps')
time.sleep(5)
```

```python
city = input('Enter City Name : ')                                         
search = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')                                                                                
time.sleep(2)
button = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]')             
button.click()                                                             
time.sleep(3)


try:
    url = driver.current_url
    print("URL Extracted: ", url)
    lat_lng = re.findall(r'@(.*)data',url)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```

“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description. 

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome()
driver.get('https://www.hostelworld.com/')
```

```python
prod_url=[]

url=driver.find_elements(By.XPATH,'//a[@class="nuxt-link"]')
for i in url:
        prod_url.append(i.get_attribute("href"))
time.sleep(3)
```

```python
len(prod_url)
```

```python
Name=[]

name=driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in name:
    Name.append(str(i.text))
Name
    
```

```python
Distance=[]

dist=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in dist:
    Distance.append(str(i.text))
Name
    
```

```python
Rating=[]
rate= driver.find_elements(By.XPATH,'//span[@class="number"]')
for i in rate:
    Rating.append(str(i.text))
Rating
```

```python
Review=[]
review=driver.find_elements(By.XPATH,'//div[@class="review"]')
for i in review:
    Review.append(str(i.text))
Review
```

```python
overall_R=[]
over_review=driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in over_review:
    overall_R.append(str(i.text))
overall_R
```

```python
Pvt_Prc=[]
pvt_prc=driver.find_elements(By.XPATH,'//div[@class="accommodation-price"]')
for i in pvt_prc:
    Pvt_Prc.append(str(i.text))
Pvt_Prc
```

```python
Dorm=[]
dorm=driver.find_elements(By.XPATH,'//div[@class="accommodation-price unavailable"]')
for i in dorm:
    Dorm.append(str(i.text))
Dorm
```

```python
discription=[]
disc=driver.find_elements(By.XPATH,'//div[@class="content"]')
for i in disc:
    discription.append(str(i.text))
discription
```

```python
Fac=[]
fac=driver.find_elements(By.XPATH,'//ul[@class="facility-sections"]')
for i in fac:
    Fac.append(str(i.text))
Fac
```
