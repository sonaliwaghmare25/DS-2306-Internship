1) Write a python program to display all the header tags from wikipedia.org and make data frame

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
heading_tags=['h1','h2']
for tags in soup.find_all(heading_tags):
    print(tags.name + '->' + tags.text.strip())
```

```python
import pandas as pd
df=pd.DataFrame({'Headers':heading_tags })
df
```

2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
sp=BeautifulSoup(page.content)
sp
```

```python
list1=[]
for i in sp.find_all('div',class_="presidentListing"):
    list1.append(i.text.split('\n')[1])
list1
```

```python
term=[]
for i in sp.find_all('div',class_="presidentListing"):
    term.append(i.text.split('\n')[2])
term
```

```python
import pandas as pd
df=pd.DataFrame({'Names':list1, 'Term of Office':term})
df
```

4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
b) Top 10 women’s ODI Batting players along with the records of their team and rating.
c) Top 10 women’s ODI all-rounder along with the records of their team and rating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')

soup
```

```python
player=[]
for i in soup.find_all('td',class_="table-body__cell name"):
    player.append(i.text)
player
```

```python
team=[]
for i in soup.find_all('td',class_="table-body__cell nationality-logo"):
    team.append(i.text)
team
```

```python
rating=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
rating
```

```python
df=pd.DataFrame({'Batting player':player,'Team':team,'Rating':rating})
df
```

5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame

i) Headline
ii) Time
iii) News Link

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
page
```

```python
sp=BeautifulSoup(page.content)
sp
```

```python
head=[]
for i in sp.find_all('div',class_="LatestNews-headlineWrapper"):
    head.append(i.text.split('|')[0])
head
```

```python
time=[]
for i in sp.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time
```

```python
link=[]
for i in sp.find_all('div',class_="SecondaryCard-imageContainer"):
    link.append(i.get)
link
```

```python
import pandas as pd
df=pd.DataFrame({'Headline':head,'Time':time})
df
```

7) Write a python program to scrape mentioned details from dineout.co.inand make data frame

 i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image UR

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
sp=BeautifulSoup(page.content)
sp
```

```python
name=[]
for i in sp.find_all('div',class_="restnt-info cursor"):
    name.append(i.text)
name
           
```

```python
cui=[]
for i in sp.find_all('span',class_="double-line-ellipsis"):
    cui.append(i.text.split('|')[1])
cui
```

```python
loc=[]
for i in sp.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc
```

```python
rate=[]
for i in sp.find_all('div',class_="restnt-rating rating-4"):
    rate.append(i.text)
rate
```

```python
image=[]
for i in sp.find_all('img',class_="no-img"):
    image.append(i.get('data-src'))
image
```

```python
print(len(name),len(cui),len(loc),len(rate),len(image))
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant Name':name,'Cuisin':cui,'Location': loc,'Ratings':rate,'Image_URL':image})
df
```
